```
import matplotlib.pyplot as plt
import numpy as np

# Define the range for x and the function y = 2x + 3
x = np.linspace(-20, 20, 500)  # Generate 500 points between -20 and 20 for a smooth graph
y = 2 * x + 3

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = 2x + 3", color='blue')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # Add y-axis
plt.title("Graph of y = 2x + 3")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Define the range for x and the function y = x^2
x = np.linspace(-10, 10, 500)
y = x**2

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='green', linestyle='--', linewidth=2, label="y = x^2")  # Customizing the plot
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # Add y-axis
plt.title("Plot of y = x^2", fontsize=14)
plt.xlabel("x-axis", fontsize=12)
plt.ylabel("y-axis", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.6)
plt.show()
```

```
import matplotlib.pyplot as plt

# Data for revenue generation of 5 different companies (in billions)
companies = ['Company A', 'Company B', 'Company C', 'Company D', 'Company E']
revenues = [45, 30, 25, 15, 35]  # Example revenue data

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    revenues,
    labels=companies,
    autopct='%1.1f%%',  # Show percentages with 1 decimal place
    startangle=140,  # Rotate the start angle of the pie chart
    colors=['gold', 'lightblue', 'lightgreen', 'pink', 'orange']  # Custom colors
)
plt.title("Revenue Generation of 5 Different Companies", fontsize=14)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Define time and cosine wave parameters
time = np.linspace(0, 10, 500)  # Time variable (0 to 10 seconds, 500 points)
frequency = 1  # Frequency of the wave (1 Hz)
amplitude = 1  # Amplitude of the wave
cos_wave = amplitude * np.cos(2 * np.pi * frequency * time)  # Cosine wave equation

# Plot the cosine wave
plt.figure(figsize=(10, 6))
plt.plot(time, cos_wave, color='blue', linestyle='-', linewidth=2, label='Cosine Wave')
plt.title("Cosine Wave Varying Over Time", fontsize=14)
plt.xlabel("Time (seconds)", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add x-axis
plt.grid(True, alpha=0.6)
plt.legend(fontsize=10)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x from 0 to 2pi
x = np.linspace(0, 2 * np.pi, 400)

# Define the trigonometric functions
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)
cot_x = 1 / np.tan(x)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot each function in a subplot
axs[0, 0].plot(x, sin_x, label='sin(x)', color='blue')
axs[0, 0].set_title('sin(x)')
axs[0, 0].grid(True)

axs[0, 1].plot(x, cos_x, label='cos(x)', color='red')
axs[0, 1].set_title('cos(x)')
axs[0, 1].grid(True)

axs[1, 0].plot(x, tan_x, label='tan(x)', color='green')
axs[1, 0].set_title('tan(x)')
axs[1, 0].set_ylim(-10, 10)  # Limit y-axis to avoid extreme values
axs[1, 0].grid(True)

axs[1, 1].plot(x, cot_x, label='cot(x)', color='purple')
axs[1, 1].set_title('cot(x)')
axs[1, 1].set_ylim(-10, 10)  # Limit y-axis to avoid extreme values
axs[1, 1].grid(True)

# Adjust layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```
